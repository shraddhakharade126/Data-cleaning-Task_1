```python
#importing liabraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read and load data to df

df=pd.read_csv("raw_messy_dataset.csv")
df
```

```python

df.info()
```

```python
#searching count of duplicates

print(df.duplicated().sum())
```

```python
# remaoving duplcates

df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python

#handling Age

df["Age"]=pd.to_numeric(df["Age"],errors="coerce")
x = df["Age"].median()
df["Age"].fillna(x, inplace = True) 
print(df.to_string())
```

```python
 #Standerized gender column using replace 
df['Gender'] = df['Gender'].replace({
     'M': 'Male', 'F': 'Female',' ':'None' 
})
print(df.to_string())
```

```python

df['Email'].fillna("-",inplace=True)
print(df.to_string())

```

```python


# from datetime import datetime

df['Join_Date'] = pd.to_datetime(df['Join_Date'], errors='coerce')

df.info()

```

```python

#handling Salary column

df["Salary"]=pd.to_numeric(df["Salary"],errors="coerce")
df.info()
```

```python
x = df["Department"].mode()[0]
x

# df["Department"].fillna(x, inplace = True) 
# print(df.to_string())
```

```python
x = df["Department"].mode()[0]

df["Department"].fillna(x, inplace = True) 
print(df.to_string())

```

```python
#Make new file of clean data

df.to_csv('Cleaned Data')
```
